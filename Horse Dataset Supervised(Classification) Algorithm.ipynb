```python
num = df[['hospital_number','rectal_temp','pulse','respiratory_rate','nasogastric_reflux_ph',
 'packed_cell_volume', 'total_protein','abdomo_protein','lesion_1', 'lesion_2','lesion_3']]
```

```python
cat = df[['surgery','age','temp_of_extremities','peripheral_pulse','mucous_membrane', 
          'capillary_refill_time','pain','peristalsis','abdominal_distention','nasogastric_tube',
          'nasogastric_reflux','rectal_exam_feces','abdomen','abdomo_appearance','surgical_lesion',
          'cp_data']]
```

```python
Y = df["outcome"]
```

```python
(df.isnull().sum()/df.shape[0])*100
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(missing_values=np.nan,strategy="mean")
```

```python
df_num_si = si.fit_transform(num)
df_num_si
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mms = MinMaxScaler()
```

```python
df_num_si_mms = mms.fit_transform(df_num_si)
df_num_si_mms
```

```python
df_num_si_mms_df = pd.DataFrame(df_num_si_mms,index=num.index,columns=num.columns)
df_num_si_mms_df
```

```python
si = SimpleImputer(missing_values=np.nan,strategy="most_frequent")
```

```python
df_cat_si = si.fit_transform(cat)
df_cat_si
```

```python
df_cat_si_df = pd.DataFrame(df_cat_si,index=cat.index,columns=cat.columns)
df_cat_si_df
```

```python
df_cat_si_df_dummies = pd.get_dummies(df_cat_si_df,drop_first=True)
df_cat_si_df_dummies
```

```python
df_final = pd.concat([df_num_si_mms_df,df_cat_si_df_dummies],axis=1)
df_final
```

```python
X = df_final
```

```python
Y = df["outcome"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Y_predict_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
LR = accuracy_score(Y_test,Y_predict_lr)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc= DecisionTreeClassifier()
```

```python
dtc.fit(X_train,Y_train)
```

```python
Y_predict_dtc = dtc.predict(X_test)
```

```python
DTC = accuracy_score(Y_test,Y_predict_dtc)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,Y_train)
```

```python
Y_predict_rfc = rfc.predict(X_test)
```

```python
RFC = accuracy_score(Y_test,Y_predict_dtc)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(X_train,Y_train)
```

```python
Y_predict_svm = svm.predict(X_test)
```

```python
SVM = accuracy_score(Y_test,Y_predict_svm)
```

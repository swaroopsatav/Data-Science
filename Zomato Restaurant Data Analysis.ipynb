# Data Analysis

```python
#Distribution of Restaurant Ratings
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=30, kde=True)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 cuisines
top_cuisines = df['Cuisines'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_cuisines.index, x=top_cuisines.values, palette='viridis')
plt.title('Top 10 Cuisines')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Top 10 cities with the most restaurants
top_cities = df['City'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=top_cities.index, x=top_cities.values, palette='viridis')
plt.title('Top 10 Cities with Most Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('City')
plt.show()
```

```python
# Average Rating by City
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(y=avg_rating_by_city.index, x=avg_rating_by_city.values, palette='viridis')
plt.title('Top 10 Cities by Average Restaurant Rating')
plt.xlabel('Average Rating')
plt.ylabel('City')
plt.show()
```

```python
# Price range distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Price range', data=df, palette='viridis')
plt.title('Price Range Distribution')
plt.xlabel('Price Range')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Correlation between Rating and Price Range
plt.figure(figsize=(10, 6))
sns.boxplot(x='Price range', y='Aggregate rating', data=df, palette='viridis')
plt.title('Rating vs Price Range')
plt.xlabel('Price Range')
plt.ylabel('Aggregate Rating')
plt.show()
```

```python
# Average cost for two people distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Average Cost for two'], bins=30, kde=True, palette='viridis')
plt.title('Distribution of Average Cost for Two People')
plt.xlabel('Average Cost for Two')
plt.ylabel('Frequency')
plt.show()
```

```python
# Most popular cuisines by city
popular_cuisines_by_city = df.groupby('City')['Cuisines'].value_counts().groupby(level=0).nlargest(1).reset_index(level=0, drop=True)
print(popular_cuisines_by_city.head(10))
```

```python
# Average ratings by cuisine
avg_ratings_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(y=avg_ratings_by_cuisine.index[:10], x=avg_ratings_by_cuisine.values[:10], palette='viridis')
plt.title('Top 10 Cuisines by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Cost for two by city
cost_for_two_by_city = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sns.barplot(y=cost_for_two_by_city.index[:10], x=cost_for_two_by_city.values[:10], palette='viridis')
plt.title('Top 10 Cities by Average Cost for Two People')
plt.xlabel('Average Cost for Two')
plt.ylabel('City')
plt.show()
```

```python
# Distribution of restaurants by cuisine type
plt.figure(figsize=(12, 8))
sns.countplot(y='Cuisines', data=df, order=df['Cuisines'].value_counts().index[:10], palette='viridis')
plt.title('Top 10 Cuisines by Number of Restaurants')
plt.xlabel('Number of Restaurants')
plt.ylabel('Cuisine')
plt.show()
```

```python
# Rating vs Price Range in different cities
cities_to_compare = df['City'].value_counts().index[:5]  # Choose top 5 cities for comparison
for city in cities_to_compare:
    plt.figure(figsize=(10, 6))
    city_data = df[df['City'] == city]
    sns.boxplot(x='Price range', y='Aggregate rating', data=city_data, palette='viridis')
    plt.title(f'Rating vs Price Range in {city}')
    plt.xlabel('Price Range')
    plt.ylabel('Aggregate Rating')
    plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Top restaurants in each city based on ratings
top_restaurants_by_city = df.loc[df.groupby('City')['Aggregate rating'].idxmax()][['City', 'Restaurant Name', 'Aggregate rating']]
print(top_restaurants_by_city)
```

```python
# Average cost for two in different price ranges
avg_cost_by_price_range = df.groupby('Price range')['Average Cost for two'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_cost_by_price_range.index, y=avg_cost_by_price_range.values, palette='viridis')
plt.title('Average Cost for Two by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Cost for Two')
plt.show()
```

```python
# Most common ratings
most_common_ratings = df['Aggregate rating'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=most_common_ratings.index, y=most_common_ratings.values, palette='viridis')
plt.title('Most Common Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```
